```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data file
data = np.loadtxt('data/rhomax.asc')
data5 = np.loadtxt('data/rhomax5.asc')
data10 = np.loadtxt('data/rhomax10.asc')
data15 = np.loadtxt('data/rhomax15.asc')

# Access a specific column
domain = data[:, 0]  # 0 refers to the first column
domain5 = data5[:, 0]
domain10 = data10[:, 0]
domain15 = data15[:, 0]
f = data[:, 1]
f5 = data5[:, 1]
f10 = data10[:, 1]
f15 = data15[:, 1]
```

```python
plt.rcParams.update({
    'axes.titlesize': 16,     # Tamao de la fuente del ttulo del eje
    'axes.labelsize': 14,     # Tamao de la fuente de las etiquetas de los ejes
    'xtick.labelsize': 12,    # Tamao de la fuente de las etiquetas del eje x
    'ytick.labelsize': 12,    # Tamao de la fuente de las etiquetas del eje y
    'legend.fontsize': 13,    # Tamao de la fuente de la leyenda
    'figure.titlesize': 18    # Tamao de la fuente del ttulo de la figura
})

plt.plot(domain, f,label=r'$\alpha = 1$')
plt.plot(domain5, f5,label=r'$\alpha = 25$')
plt.plot(domain10, f10,label=r'$\alpha = 100$')
plt.plot(domain15, f15,label=r'$\alpha = 225$')
plt.xlim(0, 130)
plt.ylim(.9, 2.2)
plt.xlabel(r'$t$')
plt.ylabel(r'$\rho_c$')
plt.savefig('rhomax.png', dpi=300, bbox_inches='tight')
# Add the legend (which shows the labels in a box)
plt.legend( loc="upper right", fontsize=10, fancybox=True)
```
